```python
# --- Cell 1: Import Libraries ---

import os
import glob
import numpy as np
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
import kagglehub  # Added kagglehub import

from scipy import stats
from scipy.signal import hilbert
from scipy.fft import fft
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, Conv1D, MaxPooling1D, LSTM,
                                     Dense, Dropout, BatchNormalization,
                                     Flatten, GlobalAveragePooling1D,
                                     Conv1DTranspose, Cropping1D)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
tf.get_logger().setLevel('ERROR')

print("All libraries imported successfully!")


# --- Cell 2: Data Loading and Exploration ---

print("Downloading Kaggle dataset using kagglehub...")
try:
    uwrfkaggler_ravdess_emotional_speech_audio_path = kagglehub.dataset_download('uwrfkaggler/ravdess-emotional-speech-audio')
    print('Data source import complete.')
except Exception as e:
    print(f"Error downloading dataset: {e}")
    print("Please ensure you are authenticated with Kaggle in this environment.")
    # Set a placeholder path to allow the script to continue and fail at glob
    uwrfkaggler_ravdess_emotional_speech_audio_path = 'uwrfkaggler/ravdess-emotional-speech-audio'

# Dataset path from kagglehub download
DATASET_PATH = uwrfkaggler_ravdess_emotional_speech_audio_path

# Glob all possible actor folders (including nested)
wav_files = glob.glob(os.path.join(DATASET_PATH, "Actor_*", "*.wav"))
wav_files += glob.glob(os.path.join(DATASET_PATH, "audio_speech_actors_01-24", "Actor_*", "*.wav"))

# Remove duplicates
wav_files = list(set(wav_files))

print(f"ğŸµ Total .wav files detected: {len(wav_files)}")

# Debugging in case
if len(wav_files) == 0:
    print(f"\nâŒ No wav files found at path: {DATASET_PATH}")
    print("Please check the downloaded dataset structure or authentication.")
    raise Exception("No wavs found. Dataset path issue or download failed.")

# Show few files to confirm
print("\nFirst 10 audio files:")
for f in wav_files[:10]:
    print(f)

# âœ… Emotion mapping function
def get_emotion_from_filename(file_path):
    file_name = os.path.basename(file_path)
    parts = file_name.split('-')
    if len(parts) > 2:
        try:
            emotion_code = int(file_name.split('-')[2])
            emotion_map = {
                1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad',
                5: 'angry', 6: 'fearful', 7: 'disgust', 8: 'surprised'
            }
            return emotion_map.get(emotion_code, 'unknown')
        except (IndexError, ValueError):
            print(f"Warning: Could not parse emotion from {file_name}")
            return 'unknown'
    print(f"Warning: Could not parse emotion from {file_name}")
    return 'unknown'

# Test emotion extraction
if wav_files:
    sample_emotions = [get_emotion_from_filename(f) for f in wav_files[:10]]
    print("\nSample emotions:", sample_emotions)
else:
    print("\nNo files to test emotion extraction.")


# --- Cell 3: Enhanced Feature Extraction with Energy, Entropy, and VMD ---

# We fix max_pad_len for consistent input shape
MAX_PAD_LEN = 130

def permutation_entropy(signal, order=3, delay=1):
    """Calculate permutation entropy efficiently"""
    n = len(signal)
    if n <= order:
        return 0

    permutations = []
    for i in range(n - delay * (order - 1)):
        # Get indices for the permutation pattern
        indices = np.argsort(signal[i:i + delay * order:delay])
        permutations.append(tuple(indices))

    if not permutations:
        return 0

    # Count unique permutations
    unique_perms, counts = np.unique(permutations, return_counts=True)
    probs = counts / len(permutations)

    # Calculate entropy
    pe = -np.sum(probs * np.log2(probs))
    return pe

def spectral_entropy(signal, sr=22050, n_fft=2048):
    """Calculate spectral entropy"""
    spectrogram = np.abs(librosa.stft(signal, n_fft=n_fft))
    psd = np.sum(spectrogram, axis=0)
    psd_norm = psd / np.sum(psd)

    # Remove zeros to avoid log(0)
    psd_norm = psd_norm[psd_norm > 0]

    if len(psd_norm) == 0:
        return 0

    spectral_ent = -np.sum(psd_norm * np.log2(psd_norm))
    return spectral_ent

def renyi_entropy(signal, alpha=2):
    """Calculate Renyi entropy"""
    # Normalize signal to probability distribution
    signal_norm = np.abs(signal) / np.sum(np.abs(signal))
    signal_norm = signal_norm[signal_norm > 0]  # Remove zeros

    if len(signal_norm) == 0:
        return 0

    if alpha == 1:
        # Shannon entropy case
        return -np.sum(signal_norm * np.log2(signal_norm))
    else:
        return (1 / (1 - alpha)) * np.log2(np.sum(signal_norm ** alpha))

def shannon_entropy(signal):
    """Calculate Shannon entropy"""
    # Convert to probability distribution
    signal_norm = np.abs(signal) / np.sum(np.abs(signal))
    signal_norm = signal_norm[signal_norm > 0]  # Remove zeros

    if len(signal_norm) == 0:
        return 0

    return -np.sum(signal_norm * np.log2(signal_norm))

def simple_vmd_features(signal, sr=22050, n_modes=3):
    """
    Simplified VMD-like feature extraction using empirical mode decomposition
    This is a computationally efficient approximation
    """
    try:
        # Use librosa's harmonic-percussive separation as a VMD approximation
        harmonic, percussive = librosa.effects.hpss(signal)

        features = []

        # Extract features from each component
        for component in [signal, harmonic, percussive]:
            if len(component) > 0:
                # Energy
                energy = np.sum(component ** 2)
                features.append(energy)

                # Spectral centroid
                centroid = librosa.feature.spectral_centroid(y=component, sr=sr)[0]
                features.append(np.mean(centroid))

                # Bandwidth
                bandwidth = librosa.feature.spectral_bandwidth(y=component, sr=sr)[0]
                features.append(np.mean(bandwidth))
            else:
                features.extend([0, 0, 0])

        return np.array(features[:9])  # Return 9 features (3 components Ã— 3 features)

    except Exception as e:
        print(f"VMD approximation error: {e}")
        return np.zeros(9)

def extract_advanced_features(data, sr=22050):
    """Extract energy, entropy, and VMD-based features"""
    features = []

    try:
        # 1. Time-domain energy features
        rms_energy = np.sqrt(np.mean(data ** 2))
        total_energy = np.sum(data ** 2)
        energy_entropy = shannon_entropy(data ** 2)  # Energy distribution entropy

        features.extend([rms_energy, total_energy, energy_entropy])

        # 2. Spectral energy features
        spectrogram = np.abs(librosa.stft(data, n_fft=1024))
        spectral_energy = np.sum(spectrogram ** 2)
        spectral_energy_bands = []

        # Divide spectrum into 4 bands and calculate energy in each
        n_bands = 4
        band_size = spectrogram.shape[0] // n_bands
        for i in range(n_bands):
            start_idx = i * band_size
            end_idx = (i + 1) * band_size if i < n_bands - 1 else spectrogram.shape[0]
            band_energy = np.sum(spectrogram[start_idx:end_idx, :] ** 2)
            spectral_energy_bands.append(band_energy)

        features.extend([spectral_energy] + spectral_energy_bands)

        # 3. Entropy features
        perm_ent = permutation_entropy(data, order=3)
        spec_ent = spectral_entropy(data, sr)
        renyi_ent = renyi_entropy(data, alpha=2)

        features.extend([perm_ent, spec_ent, renyi_ent])

        # 4. Statistical features
        skewness = stats.skew(data)
        kurtosis = stats.kurtosis(data)
        zcr = np.mean(librosa.feature.zero_crossing_rate(data)[0])

        features.extend([skewness, kurtosis, zcr])

        # 5. Simplified VMD features (computationally efficient)
        vmd_features = simple_vmd_features(data, sr)
        features.extend(vmd_features)

        # 6. Spectral rolloff and flux
        rolloff = np.mean(librosa.feature.spectral_rolloff(y=data, sr=sr)[0])
        spectral_flux = np.mean(librosa.onset.onset_strength(y=data, sr=sr))

        features.extend([rolloff, spectral_flux])

        return np.array(features)

    except Exception as e:
        print(f"Advanced feature extraction error: {e}")
        return np.zeros(29)  # Return zeros if extraction fails

def extract_features(file_path, max_pad_len=MAX_PAD_LEN):
    try:
        # Load audio file with slightly longer duration to capture more features
        data, sr = librosa.load(file_path, duration=3, offset=0.5, sr=22050)

        # Extract traditional features
        mfcc = librosa.feature.mfcc(y=data, sr=sr, n_mfcc=40, n_fft=2048, hop_length=512)
        mel_spec = librosa.feature.melspectrogram(y=data, sr=sr, n_mels=40, n_fft=2048, hop_length=512)
        chroma = librosa.feature.chroma_stft(y=data, sr=sr, n_fft=2048, hop_length=512)
        contrast = librosa.feature.spectral_contrast(y=data, sr=sr, n_fft=2048, hop_length=512)

        # Extract advanced features (energy, entropy, VMD)
        advanced_features = extract_advanced_features(data, sr)

        # Pad or truncate traditional features to fixed length
        def pad_features(features, max_len):
            if features.shape[1] < max_len:
                pad_width = max_len - features.shape[1]
                features = np.pad(features, pad_width=((0, 0), (0, pad_width)), mode='constant')
            else:
                features = features[:, :max_len]
            return features

        mfcc = pad_features(mfcc, max_pad_len)
        mel_spec = pad_features(mel_spec, max_pad_len)
        chroma = pad_features(chroma, max_pad_len)
        contrast = pad_features(contrast, max_pad_len)

        # Combine traditional features
        combined_traditional = np.vstack([mfcc, mel_spec, chroma, contrast])

        # Create advanced features matrix by repeating across time steps
        # This allows the temporal model to see these global features at each time step
        advanced_features_repeated = np.tile(advanced_features, (max_pad_len, 1))

        # Combine all features
        combined_features = np.hstack([combined_traditional.T, advanced_features_repeated])

        return combined_features  # Shape: (130, n_features)

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Test feature extraction and determine actual feature dimension
if wav_files:
    test_features = extract_features(wav_files[0])
    if test_features is not None:
        N_FEATURES = test_features.shape[1]  # Dynamically set based on actual data
        print(f"\nFeature shape: {test_features.shape}")
        print(f"Number of features per timestep: {N_FEATURES}")
    else:
        print("\nFeature extraction failed for test file")
else:
    print("\nNo audio files found. Skipping feature extraction test.")


# --- Cell 4: Process All Files and Create Dataset ---

X, y = [], []
failed_files = []

print("Extracting features from all audio files...")
for file in tqdm(wav_files):
    features = extract_features(file)
    emotion = get_emotion_from_filename(file)

    # We only include files that were processed correctly and have a known emotion
    if features is not None and emotion != 'unknown':
        X.append(features)
        y.append(emotion)
    elif features is None:
        failed_files.append(file)

print(f"\nSuccessfully processed: {len(X)} files")
print(f"Failed to process: {len(failed_files)} files")

# Check class distribution
unique, counts = np.unique(y, return_counts=True)
print("\nClass distribution:")
for emotion, count in zip(unique, counts):
    print(f"{emotion}: {count}")


# --- Cell 5: Data Preprocessing (Labels and Scaling) ---

# Convert to numpy arrays
X = np.array(X)
y = np.array(y)

print(f"Original X shape: {X.shape}")
print(f"Original y shape: {y.shape}")

# Update N_FEATURES based on actual data
N_FEATURES = X.shape[2]
print(f"Actual number of features: {N_FEATURES}")

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_categorical = to_categorical(y_encoded)
num_classes = len(le.classes_)

print(f"\nEncoded labels shape: {y_categorical.shape}")
print(f"Number of classes: {num_classes}")
print(f"Classes: {le.classes_}")

# --- Feature Scaling ---
# Reshape X to 2D for scaling: (samples * timesteps, features)
n_samples, n_timesteps, n_features = X.shape
X_reshaped = X.reshape(-1, n_features)

scaler = StandardScaler()
X_scaled_reshaped = scaler.fit_transform(X_reshaped)

# Reshape back to 3D: (samples, timesteps, features)
X_scaled = X_scaled_reshaped.reshape(n_samples, n_timesteps, n_features)

print(f"\nScaled X shape: {X_scaled.shape}")


# --- Cell 6: Build 1D Conv Autoencoder ---

# Define the input shape based on actual data
input_shape = (MAX_PAD_LEN, N_FEATURES)  # (130, actual_feature_count)

def create_autoencoder(input_shape):
    # --- Encoder ---
    inputs = Input(shape=input_shape)

    # Conv Block 1
    x = Conv1D(128, 5, padding='same', activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling1D(2, padding='same')(x) # 130 -> 65

    # Conv Block 2
    x = Conv1D(64, 3, padding='same', activation='relu')(x)
    x = BatchNormalization()(x)
    encoded = MaxPooling1D(2, padding='same', name='bottleneck')(x) # 65 -> 33
    # Bottleneck shape: (None, 33, 64)

    # --- Decoder ---
    # Decoder Block 1
    x = Conv1DTranspose(64, 3, strides=2, padding='same')(encoded) # 33 -> 66
    x = BatchNormalization()(x)

    # Decoder Block 2
    x = Conv1DTranspose(128, 5, strides=2, padding='same')(x) # 66 -> 132
    x = BatchNormalization()(x)

    # Adjust size back to 130
    # Crop 1 from the beginning and 1 from the end
    x = Cropping1D(cropping=(1, 1))(x) # 132 -> 130

    # Output layer to reconstruct original feature count
    outputs = Conv1D(input_shape[1], 1, activation='linear', padding='same')(x)  # Use actual feature count

    model = Model(inputs, outputs)
    return model

autoencoder = create_autoencoder(input_shape)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

print("Autoencoder Summary:")
autoencoder.summary()


# --- Cell 7: Train the Autoencoder ---

print("Training Autoencoder for feature representation...")

# We train the AE on the full, scaled dataset (unsupervised)
ae_history = autoencoder.fit(
    X_scaled, X_scaled,
    epochs=100, # AE can train for a good while
    batch_size=32,
    validation_split=0.1,
    callbacks=[
        EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10)
    ],
    verbose=1
)

print("âœ… Autoencoder training complete.")

# Save the encoder part for later use
encoder_model = Model(autoencoder.input, autoencoder.get_layer('bottleneck').output)
encoder_model.save('encoder_model.h5')
print("Encoder part saved as 'encoder_model.h5'")


# --- Cell 8: Build Hybrid Classifier Model (AE + CNN + LSTM) ---

def create_classifier_model(encoder_base, input_shape, num_classes):

    # Freeze the pre-trained encoder base
    encoder_base.trainable = False

    inputs = Input(shape=input_shape)

    # 1. Autoencoder Encoder Base
    x = encoder_base(inputs) # Output shape: (None, 33, 64)

    # 2. CNN Layers (to process encoded features)
    x = Conv1D(128, 3, activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    # 3. LSTM Layers (for temporal dependencies)
    x = LSTM(128, return_sequences=True, dropout=0.3)(x)
    x = LSTM(64, dropout=0.3)(x)

    # 4. Classification Head
    x = Dense(128, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)

    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)

    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)

    # Compile the model
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Test build
# We re-load the encoder to ensure it's a fresh instance for the classifier
encoder_base_test = tf.keras.models.load_model('encoder_model.h5', compile=False)
classifier_test = create_classifier_model(encoder_base_test, input_shape, num_classes)
print("\nClassifier Model Summary:")
classifier_test.summary()
del classifier_test # Clean up


# --- Cell 9: 5-Fold Cross-Validation Training ---

N_SPLITS = 5
EPOCHS = 150 # Generous epoch count, EarlyStopping will handle it
BATCH_SIZE = 32

# Use StratifiedKFold to ensure balanced class distribution in each fold
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

# Store results
all_fold_accuracies = []
class_accuracies_all_folds = []
fold_histories = []

print(f"\n--- Starting {N_SPLITS}-Fold Cross-Validation ---")

for fold, (train_index, test_index) in enumerate(skf.split(X_scaled, y_encoded)):
    fold_no = fold + 1
    print(f"\n--- FOLD {fold_no}/{N_SPLITS} ---")

    # --- 1. Split Data ---
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y_categorical[train_index], y_categorical[test_index]

    # Further split training data into train and validation (for the model.fit callback)
    X_train_fold, X_val_fold, y_train_fold, y_val_fold = train_test_split(
        X_train, y_train,
        test_size=0.15, # 15% of the training fold data for validation
        random_state=42,
        stratify=y_train.argmax(axis=1)
    )

    # --- 2. Build Model ---
    # Load a fresh, non-trained instance of the encoder for each fold
    encoder_base = tf.keras.models.load_model('encoder_model.h5', compile=False)
    model = create_classifier_model(encoder_base, input_shape, num_classes)

    # --- 3. Callbacks ---
    checkpoint_path = f"fold_{fold_no}_best_model.h5"
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=35, restore_best_weights=True), # Reduced patience
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=15, min_lr=1e-6),
        ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, mode='max')
    ]

    # --- 4. Train Model ---
    print(f"Training on {X_train_fold.shape[0]} samples, validating on {X_val_fold.shape[0]}...")
    history = model.fit(
        X_train_fold, y_train_fold,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=(X_val_fold, y_val_fold),
        callbacks=callbacks,
        verbose=1 # Set to 1 to see epoch progress
    )
    fold_histories.append(history)
    print("Training complete. Best weights restored.")

    # --- 5. Evaluate Model ---
    # Load the best weights saved by ModelCheckpoint
    model.load_weights(checkpoint_path)

    print(f"Evaluating on {X_test.shape[0]} test samples...")
    score = model.evaluate(X_test, y_test, verbose=0)
    fold_accuracy = score[1]
    all_fold_accuracies.append(fold_accuracy)
    print(f"Fold {fold_no} Test Accuracy: {fold_accuracy:.4f}")

    # --- 6. Get Class-wise Accuracy (from CM) ---
    y_pred = model.predict(X_test, verbose=0)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)

    cm = confusion_matrix(y_true_classes, y_pred_classes)
    class_accuracy = cm.diagonal() / cm.sum(axis=1)
    class_accuracies_all_folds.append(class_accuracy)

    # Clean up to save memory
    del model
    del encoder_base
    tf.keras.backend.clear_session()

print("\n--- Cross-Validation Complete ---")
print(f"Average Overall Accuracy: {np.mean(all_fold_accuracies):.4f} (+/- {np.std(all_fold_accuracies):.4f})")


# --- Cell 10: Results Visualization and Table ---

# This is needed for the 'display' function in Colab/Jupyter
try:
    from IPython.display import display
except ImportError:
    print("IPython.display not found. Skipping styled DataFrame.")
    display = print # Fallback to simple print

# Create the results table as requested
results_data = {}
class_names = le.classes_

# Populate the table with data from each fold
for i in range(N_SPLITS):
    fold_name = f"Fold {i+1}"
    fold_class_acc = class_accuracies_all_folds[i]
    results_data[fold_name] = {class_names[j]: acc for j, acc in enumerate(fold_class_acc)}
    results_data[fold_name]['Overall Accuracy'] = all_fold_accuracies[i]

# Convert to DataFrame
results_df = pd.DataFrame(results_data).T

# Calculate the average across all folds
average_accuracies = results_df.mean(axis=0)
results_df.loc['Average'] = average_accuracies

print("\n--- ğŸ“Š Final Accuracy Report ---")

# Try to display the styled DataFrame
try:
    styled_df = results_df.style.format("{:.4f}").set_properties(
        **{'border': '1px solid black', 'padding': '8px'}
    ).set_caption(
        "Cross-Validation Accuracy by Emotion"
    ).background_gradient(
        cmap='Blues', axis=1
    )
    display(styled_df)
except Exception:
    print("Could not display styled table. Printing as text.")
    print(results_df.to_markdown(floatfmt=".4f"))


# --- Plot Training Histories ---
plt.figure(figsize=(15, 6))

# Plot Accuracy
plt.subplot(1, 2, 1)
for i, history in enumerate(fold_histories):
    plt.plot(history.history['val_accuracy'], label=f'Fold {i+1} Val Acc', alpha=0.7)
plt.title('Validation Accuracy per Fold')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)

# Plot Loss
plt.subplot(1, 2, 2)
for i, history in enumerate(fold_histories):
    plt.plot(history.history['val_loss'], label=f'Fold {i+1} Val Loss', alpha=0.7)
plt.title('Validation Loss per Fold')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('cross_validation_history.png', dpi=300)
print("\nSaved cross-validation history plot to 'cross_validation_history.png'")
# plt.show() # Uncomment in an interactive environment to display plot


# --- Cell 11: Final Model Training & Inference ---

# --- Load Best Model for Inference ---
# We will use the model from the best fold (highest accuracy).

best_fold_index = np.argmax(all_fold_accuracies)
best_model_path = f"fold_{best_fold_index + 1}_best_model.h5"
print(f"\nLoading best model ({best_model_path}) for inference...")

# We need to build the model again to load the weights
encoder_base = tf.keras.models.load_model('encoder_model.h5', compile=False)
final_model = create_classifier_model(encoder_base, input_shape, num_classes)
final_model.load_weights(best_model_path)

print("âœ… Best model loaded successfully.")

# --- Inference Function ---
def predict_emotion(audio_path, model, label_encoder, scaler):
    """Predict emotion from a single audio file"""
    try:
        # 1. Extract features
        features = extract_features(audio_path) # Shape: (130, n_features)
        if features is None:
            return "Error processing audio", 0.0

        # 2. Scale features using the *same* scaler from training
        features_scaled = scaler.transform(features) # Scaler expects (n, n_features)

        # 3. Reshape for model input: (1, timesteps, features)
        features_reshaped = features_scaled[np.newaxis, ...]

        # 4. Predict
        prediction = model.predict(features_reshaped, verbose=0)
        predicted_class_index = np.argmax(prediction, axis=1)[0]
        confidence = np.max(prediction, axis=1)[0]

        # 5. Decode label
        emotion = label_encoder.inverse_transform([predicted_class_index])[0]

        return emotion, confidence

    except Exception as e:
        print(f"Inference error: {e}")
        return "Error", 0.0

# --- Test Inference ---
if wav_files:
    # Pick a random file from the original list
    test_audio_index = 150
    test_audio_file = wav_files[test_audio_index]
    actual_emotion = get_emotion_from_filename(test_audio_file)

    # Run prediction
    predicted_emotion, confidence = predict_emotion(test_audio_file, final_model, le, scaler)

    print("\n--- ğŸ™ï¸ Inference Test ---")
    print(f"Audio file: {os.path.basename(test_audio_file)}")
    print(f"Actual emotion:    {actual_emotion}")
    print(f"Predicted emotion: {predicted_emotion}")
    print(f"Confidence:        {confidence:.4f}")
else:
    print("\nNo audio files found. Skipping inference test.")

print("\n--- SCRIPT COMPLETE ---")
```
